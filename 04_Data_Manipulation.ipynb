### Adding colomn

```python
import pandas as pd 
import numpy as np

customer_data = {
    'CustomerID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'Diana Miller', 'Eve Davis', 'Frank Green', 'Grace Hall', 'Henry King'],
    'Age': [34, 45, 29, 52, 38, 30, 25, 48],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Hyderabad', 'Pune', 'Kolkata', 'Delhi'],
    'Last_Purchase_Amount': [150.75, 200.00, 50.25, 300.50, 120.00, 85.00, 10.00, 250.00],
    'Is_Premium_Customer': [True, False, False, True, False, False, True, True],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'diana@example.com',
              np.nan, 'frank@example.com', 'grace@example.com',np.nan]  
}
df_customer = pd.DataFrame(customer_data)
```

```python
df_customer['country'] = ['India', 'nijeriya', 'japan', 'thai', 'honag' , 'china','pkistan', 'afgnistan']
# print(df_customer)

# Adding colomn based on existing colomn 
df_customer['reservation'] = [True if g == 'Female' else False for g in df_customer['Gender']]
print(df_customer)
```

### removing colomns

```python
# df_customer.drop('country' , axis= 1, inplace=True)
# print(df_customer)

# removing row 
df_customer.drop([6,7], axis=0 , inplace=True)
print(df_customer)
```

### Handling Missing Values (NaN)

```python
df_customer.isnull() # data frame with flase if thereis data and Nan == true 
df_customer.notnull()  # a data frame with true if there is not Nan
```

```python
df_customer.fillna('@mail.com')
df_customer.dropna( axis=1 )  
```

### Sorting Data 

```python
df_customer.sort_values(by='Age')
df_customer.sort_values(by=['Age','Name'] , ascending=[True , False] , inplace= True)
print(df_customer) 
```

### Applying Simple Functions (apply(), map())

```python
s = df_customer['Gender'].apply(lambda g :  g.upper())
print(s)
s = df_customer['Gender'].apply(lambda g : "Neutral")
print(s)
```
